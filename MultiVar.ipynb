```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df = pd.read_csv("/kaggle/input/student-marks-dataset/Student_Marks.csv")
print(df.head())
```

```python
input_colums = df[['number_courses', 'time_study']]
X = input_colums.to_numpy()
print(X.shape)
```

```python
output_colums = df[['Marks']]
Y = output_colums.to_numpy()
```

```python
# Assuming an 80-20 train-test split
split_index = int(0.8 * len(X))
X, X_test = X[:split_index], X[split_index:]
Y, Y_test = Y[:split_index], Y[split_index:]
```

```python
w = np.ones(X.shape[1])
b = 0
```

```python
def f(X, w, b):
    return (np.dot(X, w) + b).reshape(X.shape[0], 1)
```

```python
def cost(X, Y, w, b):
    return np.mean((f(X, w, b).T - Y) ** 2) * 0.5

print(Y.shape)
```

```python
def gradient_descent(X, Y, w, b, alpha):
    w_change = np.mean((f(X, w, b) - Y) * X, axis=0)
    b_change = np.mean((f(X, w, b) - Y))
    
    return (w - (alpha * w_change), b - (alpha * b_change))
```

```python
alpha = 0.00001
prev_cost = float('inf')  # Initialize previous cost with infinity
cost_history = []  # List to store cost values for each iteration
w_history = []  # List to store w values for each iteration
b_history = []  # List to store b values for each iteration

for i in range(1000):  # Limiting to 100 iterations for safety
    current_cost = cost(X, Y, w, b)
    cost_history.append(current_cost)  # Store current cost
    w_history.append(w)  # Store current w
    b_history.append(b)  # Store current b
    
    if abs(current_cost - prev_cost) < 0.0001:
        print("Convergence reached. Stopping the optimization.")
        break
    
    w, b = gradient_descent(X, Y, w, b, alpha)
    prev_cost = current_cost
print(current_cost, i)
```

```python
print(cost(X_test, Y_test, w, b))
print(w, b)
```

Using SVD

```python
# append a columns of 1s (these are the biases)
A = X.copy()
A = np.column_stack([A, np.ones(A.shape[0])])
```

```python
U,S,Vt = np.linalg.svd(A, full_matrices=False)
```

```python
# solve Ax = b for the best possible approximate solution in terms of least squares
x_hat = Vt.T @ np.linalg.inv(np.diag(S)) @ U.T @ Y
print(x_hat)
```

```python
w.shape
print(cost(X_test, Y_test, x_hat[0:w.shape[0]], x_hat[-1]))
print(x_hat[0:w.shape[0]])
print(x_hat[-1])
```
